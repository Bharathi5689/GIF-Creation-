```python
import os
from IPython.display import display, Image

# Directory containing your images
image_path = './data'

# Iterate through the files in the directory and display images
for filename in os.listdir(image_path):
    
    # Check if the file has a valid image extension
    if filename.endswith(('.jpg', '.png', '.jpeg')):
        
         # Construct the full file path by joining the directory path and the current filename
        filepath = os.path.join(image_path, filename)
        
        # Display the images in filepath
        display(Image(filename = filepath, width = 250))
```

```python
from moviepy.editor import ImageSequenceClip
import os

# Initialize an empty list to store the file paths of images
image_list = []

# Specify the path to the directory containing images
image_path = "./data"

# Set the duration for each frame in seconds
frame_duration = 2.0

# Iterate over each file in the specified directory
for filename in os.listdir(image_path):
    
    # Check if the file has a valid image extension
    if filename.endswith(('.png', '.jpeg', '.jpg')):
        
        # Construct the full file path by joining the directory path and the current filename
        file_path = os.path.join(image_path, filename)
        # Append the file path to the list
        image_list.append(file_path)
        
# Calculate the total duration of the video based on the frame duration and the number of frames
duration = frame_duration * len(image_list)

# Create an ImageSequenceClip from the list of image file paths
clip = ImageSequenceClip(image_list, fps = 1, durations = duration)

# Write the sequence of images to a GIF file
clip.write_gif("image.gif")

# Display the GIF in the IPython environment with a specified width
clip.ipython_display(width = 250, fps = 1, duration = duration)

```
